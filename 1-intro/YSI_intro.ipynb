## Getting data

There are many ways to get data:  

### Data dumps

We download a file (from the web, USB stick etc.)
 
 

### Databases

We write a SQL query. e.g.

```
SELECT Name, Number_Of_Hands FROM Characters WHERE
Name LIKE '%Skywalker%';
```
Result:
``` 
| Name             | Number_Of_Hands   |
|------------------+-------------------|
| Luke Skywalker   |                 1 |
| Anakin Skywalker |                 1 |
| Leia Skywalker   |                 2 |
| Shmi Skywalker   |                 2 |
```

### **API's** - "Application Programming Interface"



![alt](figures/collie.png)

[Tutorial:](https://colab.research.google.com/github/nestauk/im-tutorials/blob/3-ysi-tutorial/notebooks/APIs/API_tutorial.ipynb)
https://bit.ly/2kqXHKF

### **Web scraping**

![](figures/webscraping.jpg)

[Tutorial:](https://colab.research.google.com/github/nestauk/im-tutorials/blob/3-ysi-tutorial/notebooks/Web-Scraping/Web%20Scraping%20Tutorial.ipynb)
https://bit.ly/2lTaqGd
